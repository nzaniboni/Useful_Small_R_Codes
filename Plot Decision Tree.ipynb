### Plotting Decision Trees

```r
#Create a decision tree
data(iris)
library(rpart)
arvore <- rpart(Species ~ ., method="class", data=iris)
```

```r
#Presenting the rules
arvore
```

```r
#First way to plot the tree
library(rpart.plot)
rpart.plot(arvore, type = 4, extra = 105, cex=1)
```

```r
#Second way to plot the tree
library(rattle)
fancyRpartPlot(arvore)
```
